# Categorizing data & Creating Features 

```python
columns = [
    'father_qualification',
    'father_occupation',
    'mother_qualification',
    'mother_occupation'
]

for col in columns:
    # Get unique values and sort them numerically
    unique_vals = np.sort(df[col].unique())
    
    print(f"\n{col} ({len(unique_vals)} unique values):")
    print(unique_vals)
```

```python
qualification_map = {
    1: 'No Formal Education',
    2: 'Primary',
    3: 'SSC',
    4: 'HSC',
    5: 'Diploma / Vocational',
    6: 'Graduate',
    7: 'Postgraduate',
    88: 'Jobless'
}

occupation_map = {
    1: 'Unemployed',
    2: 'Student',
    3: 'Agriculture / Farming',
    4: 'Business / Self-employed',
    5: 'Government Service',
    6: 'Private Service',
    7: 'Teaching / Education',
    8: 'Health Services',
    9: 'Technical / IT',
    10: 'Labour / Daily Wage',
    11: 'Police / Armed Forces',
    12: 'Retired',
    13: 'Other Occupations',
    14: 'Freelance / Contractual',
    15: 'Homemaker',
    88: 'Jobless'
}


df["father_qualification_category"] = df["father_qualification"].map(qualification_map)
df["mother_qualification_category"] = df["mother_qualification"].map(qualification_map)

df["father_occupation_category"] = df["father_occupation"].map(occupation_map)
df["mother_occupation_category"] = df["mother_occupation"].map(occupation_map)
```

```python
# Socio-Economic Status based occupation groups

def occupation_group(occupation):
    if occupation in ['Unemployed', 'Jobless', 'Student', 'Homemaker', 'Retired']:
        return 'Not Economically Active'
    elif occupation in ['Labour / Daily Wage', 'Agriculture / Farming']:
        return 'Low Stability'
    elif occupation in ['Business / Self-employed', 'Freelance / Contractual']:
        return 'Medium Stability'
    elif occupation in ['Government Service', 'Private Service', 
                        'Teaching / Education', 'Health Services', 
                        'Technical / IT', 'Police / Armed Forces']:
        return 'High Stability'
    else:
        return 'Other'


df['father_occupation_group'] = df['father_occupation_category'].apply(occupation_group)
df['mother_occupation_group'] = df['mother_occupation_category'].apply(occupation_group)
```

```python
def health_category_by_BMI(row):
    age = row['age']
    bmi = row['BMI']
    sex = row['sex_label']  # 'Male' or 'Female'
    
    if sex == 'Male':
        if 14 <= age <= 15:
            if bmi < 18.5: return 'Underweight'
            elif bmi <= 24.9: return 'Healthy'
            elif bmi <= 29.9: return 'Overweight'
            else: return 'Obese'
        elif 16 <= age <= 17:
            if bmi < 18.5: return 'Underweight'
            elif bmi <= 24.9: return 'Healthy'
            elif bmi <= 29.9: return 'Overweight'
            else: return 'Obese'
        elif 18 <= age <= 19:
            if bmi < 18.5: return 'Underweight'
            elif bmi <= 24.9: return 'Healthy'
            elif bmi <= 29.9: return 'Overweight'
            else: return 'Obese'
    else:  # Female
        if 14 <= age <= 15:
            if bmi < 18.5: return 'Underweight'
            elif bmi <= 22.9: return 'Healthy'
            elif bmi <= 28.9: return 'Overweight'
            else: return 'Obese'
        elif 16 <= age <= 17:
            if bmi < 18.5: return 'Underweight'
            elif bmi <= 23.9: return 'Healthy'
            elif bmi <= 28.9: return 'Overweight'
            else: return 'Obese'
        elif 18 <= age <= 19:
            if bmi < 18.5: return 'Underweight'
            elif bmi <= 22.9: return 'Healthy'
            elif bmi <= 28.9: return 'Overweight'
            else: return 'Obese'

df['health_category_by_BMI'] = df.apply(health_category_by_BMI, axis=1)
```

## Other Health Features

```python
# Mean Arterial Pressure (MAP)
df["Mean Arterial Pressure (MAP)"] = (2 * df["bp_dia"] + df["bp_sys"]) / 3
```

```python
# Pulse Pressure (PP)
df["pulse_pressure"] = df["bp_sys"] - df["bp_dia"]
```

```python
# Blood Pressure Category
def bp_category(row):
    sys = row["bp_sys"]
    dia = row["bp_dia"]

    if sys < 120 and dia <80:
        return "Normal"
    elif 120 <= sys < 130 and dia < 80:
        return "Elevated"
    elif 130 <= sys < 140 or 80 <= dia < 90:
        return "Hypertension Stage 1"
    elif sys >= 140 or dia >= 90:
        return "Hypertension Stage 2"
    else:
        return "Unknown"

df["bp_category"] = df.apply(bp_category, axis=1)
```

```python
# Pulse Status
def pulse_status_adolescent(row):
    age = row["age"]
    sex = row["sex_label"]
    pulse = row["pulse"]

    if sex == "Male":
        if 14 <= age <= 15:
            return "Bradycardia" if pulse < 60 else "Tachycardia" if pulse > 95 else "Normal"
        elif 16 <= age <= 17:
            return "Bradycardia" if pulse < 60 else "Tachycardia" if pulse > 90 else "Normal"
        elif 18 <= age <= 19:
            return "Bradycardia" if pulse < 60 else "Tachycardia" if pulse > 85 else "Normal"

    else:    # "Female"
        if 14 <= age <= 15:
            return "Bradycardia" if pulse < 65 else "Tachycardia" if pulse > 100 else "Normal"
        elif 16 <= age <= 17:
            return "Bradycardia" if pulse < 65 else "Tachycardia" if pulse > 95 else "Normal"
        elif 18 <= age <= 19:
            return "Bradycardia" if pulse < 65 else "Tachycardia" if pulse > 95 else "Normal"


df["pulse_status"] = df.apply(pulse_status_adolescent, axis=1)
```

```python
# Cardiovascular Risk Flag 
def cardio_risk(row):
    if row['bp_category'] in ['Hypertension Stage 1', 'Hypertension Stage 2'] or \
       row['pulse_status'] in ['Bradycardia', 'Tachycardia']:
        return 'At Risk'
    else:
        return 'Stable'

df['cardio_risk'] = df.apply(cardio_risk, axis=1)
```

## SWEMWS, AIS, FIES, stress, Depression, Anxiety

```python
for col in [
    'categorical_SWEMWS',
    'categorical_AIS',
    'categorical_stress',
    'categorical_Depression',
    'categorical_Anxiety'
]:
    print(df[col].value_counts().sort_index())
    print('-'*40)
```

```python
# SWEMWS (Well-being)
swemws_map = {
    1: 'Low Wellbeing',
    2: 'Moderate Wellbeing',
    3: 'High Wellbeing'
}
df['SWEMWS_category'] = df['categorical_SWEMWS'].map(swemws_map)
```

```python
# AIS (Insomnia)
ais_map = {
    1: 'No Insomnia',
    2: 'Moderate Insomnia',
    3: 'Severe Insomnia'
}
df['AIS_category'] = df['categorical_AIS'].map(ais_map)
```

```python
# Stress
stress_map = {
    1: 'No Stress',
    2: 'Mild Stress',
    3: 'Moderate Stress',
    4: 'High Stress',
    5: 'Severe Stress'
}
df['Stress_category'] = df['categorical_stress'].map(stress_map)
```

```python
# Depression
depression_map = {
    1: 'No Depression',
    2: 'Mild Depression',
    3: 'Moderate Depression',
    4: 'High Depression',
    5: 'Severe Depression'
}
df['Depression_category'] = df['categorical_Depression'].map(depression_map)
```

```python
# Anxiety
anxiety_map = {
    1: 'No Anxiety',
    2: 'Mild Anxiety',
    3: 'Moderate Anxiety',
    4: 'High Anxiety',
    5: 'Severe Anxiety'
}
df['Anxiety_category'] = df['categorical_Anxiety'].map(anxiety_map)
```

```python
# Fix SWEMWS direction 
# Because SWEMWS runs opposite, create a risk-aligned version.
# Now all mental health indicators point in the same direction.

risk_level = {
    1: 'High Risk',
    2: 'Moderate Risk',
    3: 'Low Risk'
}

df['SWEMWS_risk_level'] = df['categorical_SWEMWS'].map(risk_level)
```

```python
# Create aligned “risk” flags

df['stress_risk'] = df['categorical_stress'].apply(lambda x: 'At Risk' if x >= 3 else 'Low Risk') 
df['depression_risk'] = df['categorical_Depression'].apply(lambda x: 'At Risk' if x >= 3 else 'Low Risk') 
df['anxiety_risk'] = df['categorical_Anxiety'].apply(lambda x: 'At Risk' if x >= 3 else 'Low Risk') 
df['insomnia_risk'] = df['categorical_AIS'].apply(lambda x: 'At Risk' if x >= 2 else 'Low Risk')
```

```python
# Create a unified mental health risk flag

    # Checks all four domains
    # If any one domain is at risk → At Risk
    # Else → Low Risk

df['any_mental_health_risk'] = df.apply(
    lambda row: 'At Risk' if (
        row['categorical_stress'] >= 3 or
        row['categorical_Depression'] >= 3 or
        row['categorical_Anxiety'] >= 3 or
        row['categorical_AIS'] >= 2
    ) else 'Low Risk',
    axis=1
)
```
